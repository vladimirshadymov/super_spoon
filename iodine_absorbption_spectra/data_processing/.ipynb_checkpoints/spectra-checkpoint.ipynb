# Спектры поглощения йода

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
from matplotlib.ticker import MultipleLocator
```

```python
def get_spectrum(filename):
    df = pd.read_csv(filename, sep=' ').iloc[:,:2].copy()
    df = df.rename(index=str, columns={'Wavelength': 'Wavelength, nm', 'Trans': 'Transmittance, %'})
    df = df.iloc[1:,:].copy()
    df = df.astype('float', copy=True)
    return df.copy()
```

```python
df30 = get_spectrum('30_480_655.dat')
df30
```

```python
df50 = get_spectrum('50_480_655.dat')
df50
```

```python
df70 = get_spectrum('70_480_655_01.dat')
df70
```

```python
df70.iloc[394:414,:]
```

```python
from scipy.signal import find_peaks
peaks, _ = find_peaks(-df70.iloc[:,1][:800], distance=10, width=1, prominence=0.1)
peaks = np.append(peaks,401)
peaks.sort()
peaks
```

```python
fig = plt.figure()
ax = plt.subplot(111)
df30.plot(x='Wavelength, nm', y='Transmittance, %', ax=ax, label='30$^0C$')
df50.plot(x='Wavelength, nm', y='Transmittance, %', ax=ax, label='50$^0C$')
df70.plot(x='Wavelength, nm', y='Transmittance, %', ax=ax, label='70$^0C$')
ax.set_xlabel('Длина волны, нм', size=20)
ax.set_ylabel('Прозрачность, %', size=20)
ax.set_title("Прозрачность $I_2$ при ширине щели $0.1$ нм", size=20)
ax.axvline(x=559.5, linewidth=1, ymin=0.05, ymax=0.95, linestyle='-', color='r', label='Граница электронного перехода')

xminorLocator = MultipleLocator(1)
yminorLocator = MultipleLocator(1)
xmajorLocator = MultipleLocator(10)
ymajorLocator = MultipleLocator(5)

ax.xaxis.set_minor_locator(xminorLocator)
ax.yaxis.set_minor_locator(yminorLocator)
ax.xaxis.set_major_locator(xmajorLocator)
ax.yaxis.set_major_locator(ymajorLocator)
ax.set_ylim(7,78)

plt.grid(which='major', ls='-', lw=0.5, c='k')
plt.grid(which='minor', ls='--', lw=0.5, c='grey')
ax.legend(fontsize='xx-large')
j = 1
p = 0
peaks_len = len(peaks)
for i in range(1, peaks_len-14):
    if j>22 and p==0: 
        j = 20
        p = 1
    ax.annotate('%d'%j, xy=(df70.iloc[peaks[peaks_len-i],0], df70.iloc[peaks[peaks_len-i],1]-.5), 
                xytext=(df70.iloc[peaks[peaks_len-i],0]+0.8, df70.iloc[peaks[peaks_len-i],1]-5), size=15, 
                arrowprops=dict(facecolor='black', shrink=0.01, width=0.5, headwidth=5))
    j+=1

fig.set_size_inches(25.5, 12.5)
fig.savefig('trans_357_spectrum.png', dpi=300)
```
